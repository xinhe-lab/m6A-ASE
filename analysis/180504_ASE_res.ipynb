```python
import rpy2.robjects as robjects
from rpy2.robjects import pandas2ri
pandas2ri.activate()
import os
import pandas as pd
from collections import OrderedDict
path = os.path.expanduser("~/Documents/m6A")
readRDS = robjects.r['readRDS']
```

```python
def RDS2df(RBP):
    df = readRDS(f'{path}/Data/ASE/samples/{RBP}.wasp_remapped.qual10.dedup.quasar.ase.rds')
    df = pandas2ri.ri2py(df)
    tmp = df[0][0]
    print (type(tmp[0]))
    RBP_df = pd.DataFrame(data = OrderedDict({"rsID": tmp[0], "chr": list(tmp[1]), "pos": list(tmp[2]), 
                                              "beta": list(tmp[3]), "se": list(tmp[4]), "pval": list(tmp[5])}))
    return RBP_df
```
